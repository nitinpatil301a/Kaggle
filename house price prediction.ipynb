```python
df['LotFrontage']=df['LotFrontage'].fillna(df['LotFrontage'].mean())

mode_features=['BsmtCond','BsmtQual','FireplaceQu','GarageType','GarageYrBlt','GarageFinish','GarageQual','GarageCond',
               'BsmtExposure','BsmtFinType1','BsmtFinType2','MasVnrType','MasVnrArea','Electrical','GarageCars','GarageArea',
               'BsmtFullBath','BsmtHalfBath','TotalBsmtSF','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','MSZoning','Utilities',
               'Exterior1st','Exterior2nd','KitchenQual','Functional','SaleType'
              ]

def fill_by_mode(features):
    global df
    for feature in features:
        df[feature]=df[feature].fillna(df[feature].mode()[0])
        

fill_by_mode(mode_features)
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

dummy_features=['Street','MSZoning','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood',
                'Condition2','HouseStyle','RoofStyle','Exterior1st','Exterior2nd','MasVnrType',
                'ExterQual','ExterCond','Foundation','BsmtQual','CentralAir','Electrical','FireplaceQu',
                'GarageType','GarageFinish','BsmtExposure','BsmtFinType1','HeatingQC',
                'KitchenQual','GarageQual','PavedDrive','SaleType','SaleCondition',
                'BldgType','BsmtCond','BsmtFinType2','Condition1','Functional','GarageCond','Heating','RoofMatl'
               ]

def create_dummy_features(features):
    global df
    for feature in features:
        df=pd.concat([df,pd.get_dummies(df[feature],drop_first=True,prefix=feature)],axis=1)
        df.drop([feature],axis=1,inplace=True)
        #df[feature]=le.fit_transform(df[feature])
        

create_dummy_features(dummy_features)


```

```python
df.drop(['Id'],axis=1,inplace=True)
df_train=df.iloc[0:1460,]
df_test=df.iloc[1460:,]
df_test=df_test.drop(['SalePrice'],axis=1)
x_train=df_train.drop(['SalePrice'],axis=1)
y_train=df_train['SalePrice']
```

```python
df_train['SalePrice']
```

```python
from sklearn.linear_model import LinearRegression ,Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
import statsmodels.api as sm
from sklearn.model_selection import KFold,cross_val_score,cross_validate
from sklearn.model_selection import GridSearchCV

```

```python
def eval_model(models):
    for model in models:
        clf = model[1]
        print(model[0])
        kfold = KFold(n_splits=10)
        results = cross_val_score(clf, x_train, y_train, cv=kfold,scoring="r2")
        print(results.mean())
```

```python
models = []
models.append(('Linear Reg', LinearRegression()))
models.append(('Decision Tree', DecisionTreeRegressor()))
models.append(('Random Forest', RandomForestRegressor()))
models.append(("XGboost",XGBRegressor(learning_rate =0.05, n_estimators=100,gamma=0, subsample=0.8, colsample_bytree=0.8,
 nthread=4, scale_pos_weight=1, max_depth= 6, min_child_weight= 4)))
models.append(('Lasso',Lasso()))
eval_model(models)
```

```python
regressor=XGBRegressor(learning_rate =0.05, n_estimators=100,gamma=0, subsample=0.8, colsample_bytree=0.8,
 nthread=4, scale_pos_weight=1, max_depth= 6, min_child_weight= 4)
regressor.fit(x_train, y_train)
```

```python
y_pred=regressor.predict(df_test)
df_test.head()
```

```python
y_pred=pd.DataFrame(regressor.predict(df_test))
```

```python
submit=pd.read_csv("sample_submission.csv")
result=pd.concat([submit['Id'],y_pred],axis=1)
result.columns=['Id','SalePrice']
result.to_csv("results.csv",index=False)
```
